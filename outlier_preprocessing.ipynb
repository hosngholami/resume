## find and replace avg outliers metodes

```python
def find_avg_outliers(data):

    max_avg = 20
    min_avg = 0

    outlier_avg = []
    for item in data['معدل']:
        if (item > max_avg) or (item < min_avg):
            outlier_avg.append(item)

    return outlier_avg


# find_avg_outliers(data)
```

```python
def find_avg_mean(data, outliers):

    sum_avg = 0
    count_avg = 0
    for item in data['معدل']:
        if item not in outliers:
            sum_avg += item
            count_avg += 1

    return int(sum_avg / count_avg)
```

```python
def handle_avg_outliers(data, outliers, avg_mean):

    len_data = len(data['معدل'])
    new_avgs = [0] * len_data
    for i in range(len_data):
        if data.iloc[i]['معدل'] in outliers:
            new_avgs[i] = avg_mean
        else:
            new_avgs[i] = data.iloc[i]['معدل']

    data['معدل'] = new_avgs
    return data
```

## find and replace avg outliers with mean

```python
outliers = find_avg_outliers(data)
avg_mean = find_avg_mean(data, outliers)
data = handle_avg_outliers(data, outliers, avg_mean)
data['معدل']
```

## find and replace other columns outliers

### using Interquartile Range

```python
import numpy as np
import seaborn as sns
```

```python
def find_outliers(data):

    int_data = []
    for item in data:
        int_data.append(int(item))

    sort_data = np.sort(int_data)
    # sns.boxplot(sort_data)
   
    Q1 = np.percentile(int_data, 25, method = 'midpoint')
    Q2 = np.percentile(int_data, 50, method = 'midpoint') 
    Q3 = np.percentile(int_data, 75, method = 'midpoint')
    IQR = Q3 - Q1
    
    low_lim =  Q1 - 1.5 * IQR
    up_lim = Q3 + 1.5 * IQR
   
    outliers = []
    for item in int_data:
        if (item > up_lim) or (item < low_lim):
            outliers.append(item) 
   
    return outliers
    
# there is not any outlier in age column
```

```python
def cal_mean(data, outliers):
    
    non_outliers_sum = 0
    non_outliers_count = 0
    for item in data:
        if item not in outliers:
            non_outliers_sum += item
            non_outliers_count += 1

    return int(non_outliers_sum / non_outliers_count)
```

```python
def handle_column_utliers(data, outliers, column_mean):

    len_data = len(data)
    new_data = [0] * len_data
    for i in range(len_data):
        if data[i] in outliers:
            new_data[i] = column_mean
        else:
            new_data[i] = data[i]

    return new_data
    
```

```python
def handle_numeric_outliers(data, columns):

    for column in columns:

        column_data = data[column]
        column_outliers = find_outliers(column_data)
        column_mean = cal_mean(column_data, column_outliers)
        new_column_data = handle_column_utliers(column_data, column_outliers, column_mean)
        data[column] = new_column_data
```

```python
numeric_columns = ['سن', 'تجربه کاری(به سال)']
handle_numeric_outliers(data, numeric_columns)
data
```
